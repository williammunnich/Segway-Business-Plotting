```python
import pandas as pd
import datetime

def parse_date(date_str):
    """Parse date strings into datetime objects or None for invalid dates."""
    if pd.isnull(date_str):
        return None
    try:
        return datetime.datetime.strptime(date_str.split(":")[-1].strip(), "%m/%d/%y")
    except ValueError:
        return None

def prepare_data(csv_file):
    """Read the CSV file and prepare plot point data."""
    # Read the CSV file
    data = pd.read_csv(csv_file)
    
    # Parse dates
    data['Date Added'] = data['Date Added'].apply(parse_date)
    data['Date Ended'] = data['Date Ended'].apply(parse_date)
    
    # Map locations to coordinates
    locations = {
        "Anchorage, Alaska": (61.2175, -149.8949),
        "Paris, France": (48.8566, 2.3522),
        "Los Angeles, California": (34.0522, -118.2437),
        "Sarasota, Florida": (27.3364, -82.5307),
    }
    data['Coordinates'] = data['Location'].map(locations)
    
    # Prepare a list of plot points
    plot_points = []
    for _, row in data.iterrows():
        if pd.isnull(row['Coordinates']):
            continue
        plot_points.append({
            'title': row['Title'],
            'coordinates': row['Coordinates'],
            'date_added': row['Date Added'],
            'date_ended': row['Date Ended'],
        })
    
    return plot_points
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

def plot_map(plot_points, current_date):
    """Plot points on a map based on their active or inactive status."""
    # Create a map with Plate Carree projection
    plt.figure(figsize=(12, 6))
    ax = plt.axes(projection=ccrs.PlateCarree())
    ax.set_global()
    ax.coastlines()
    ax.gridlines(draw_labels=True)
    ax.set_title(f"World Map - {current_date.strftime('%B %Y')}", fontsize=14)

    # Plot each point
    for point in plot_points:
        lat, lon = point['coordinates']
        date_added = point['date_added']
        date_ended = point['date_ended']
        
        if date_added and date_added <= current_date:
            if date_ended and date_ended <= current_date:
                color = 'red'  # Inactive
            else:
                color = 'blue'  # Active
            ax.plot(lon, lat, 'o', color=color, markersize=8, transform=ccrs.PlateCarree())
            ax.text(lon + 3, lat, point['title'], color=color, alpha=0.8, transform=ccrs.PlateCarree())
    
    plt.show()
```

```python
csv_file = 'scraped_data/scraped_tour_info_20241117_212053.csv' 
plot_points = prepare_data(csv_file)
```

```python
# Example: Plotting for November 2023
current_date = datetime.datetime(2021, 11, 1)
plot_map(plot_points, current_date)
```
